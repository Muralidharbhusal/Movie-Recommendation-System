## Cleaning the data
As we can see, the data provided in the dataframe is a list of dictionaries. In the first step we attempt to convert the list of dictionaries into list of strings.

```python
# Create a python function that converts the list of dictionaries into list of strings
def converter(obj):
    A = []
    for i in ast.literal_eval(obj):
        A.append(i["name"])
    return A
```

```python
# Apply the function in the required colmns of the dataframe
# First we apply it in 'genres
movies['genres'] = movies['genres'].apply(converter)
```

```python
# We need to extract the crew as well. But here, we will only extract the director for now. 
# For that, we have to write a python function
def fetch_director(obj):
    A = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            A.append(i['name'])
            break
    return A
```

```python
# Apply the fnction in the crew column
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# The overview column is different than the other columns here. We convert the column into list as well using a lambda function
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

## Remove the Spaces between the words
To make the data more readble by the model and avoid errors, we have to remove the spaces between the words.


```python
# This function has to remove space between names and surnames
# For example "Sam Worthington" chnages into "SamWorthington"
# Same applies for other coumns as well
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "")for i in x])
```

```python
# Check if it worked
movies.head(2)
```

## Create a different column with all the necessary data obtained so far

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew']
movies
```

```python
# Now that we have created a 'tags' column with necessary data, we can drop other columnsto create a new df
movies_df = movies[['movie_id','title','tags']]
```

```python
# Convert the list of strings into strings
movies_df['tags'] = movies_df['tags'].apply(lambda x: " ".join(x))
```

```python
# Convert all the words in the 'tags' columns into lowercase
movies_df['tags'] = movies_df['tags'].apply(lambda x:x.lower())
```

## Vectorization and similarity score


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words='english')

vectors = cv.fit_transform(movies_df['tags']).toarray()

vectors
```

## Stemming
Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma.

```python
# For this we need nltk
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Write a function for stemming
def stem(text):
    A = []
    for i in text.split():
        A.append (ps.stem(i))
    return " ".join(A)
```

```python
# Apply the function
movies_df['tags'] =  movies_df['tags'].apply(stem)
```

```python
# Check 
movies_df['tags'][0]
```

```python
# Cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
similarity
```

```python
similarity[0]
```

```python
# Write a function to recommend the movies
def recommend(movie):
    movie_name = movies_df[movies_df['title']==movie].index[0]
    distances = similarity[movie_name]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(i[0])
```

```python
# Check
recommend('Avatar')
```
